```python
from pyparsing import *
import main_types
import unittest
import collections
import xml.etree.ElementTree as ET
from xmltodict import *


''' Qualifier Type '''
class Qualifier:
    def __init__(self):
        pass

class Linear(Qualifier):
    def __init__(self):
        pass

class Unlimited(Qualifier):
    def __init__(self):
        pass

''' Var Type '''
Var = str

''' Ty Type '''
class Ty :
    def __init__(self):
        pass
        
class TyBool(Ty):
        def __init__(self):
            pass
            
class TyEnd(Ty):
        def __init__(self):
            pass
            
class TyQual(Ty): 
    def __init__(self, qual,pre):
        self.qual = qual
        self.pre = pre

''' PreType Class '''
class PreType:
    def __init__(self):
        pass

class TyRecv(PreType):
    def __init__(self, ty1, ty2):
        self.ty1 = ty1
        self.ty2 = ty2
        
class TySend(PreType):
    def __init__(self, ty1, ty2):
        self.ty1 = ty1
        self.ty2 = ty2

''' Value Type'''
class Value:
    def __init__(self):
        pass

class Variable(Value):
    def __init__(self, var):
        self.val = var
    
class VBool(Value):
    def __init__(self, boolval):
        self.boolval = boolval

''' Channel Type '''
class Channel:
    def __init__(self):
        pass

''' Process Type '''
class Process:
    def __init__(self):
        pass

class Output(Process):
    def __init__(self):
        pass

class Input(Process):
    def ___init___(self):
        pass
        
class Par(Process):
    def __init__(self):
        pass

class If(Process):
    def __init__(self):
        pass
        
class Inaction(Process):
    def __init__(self):
        pass

class ScopeRestriction(Process):
    def __init__(self):
        pass


linear = Group(Literal('lin'))("LinearNode")
unrestricted = Group(Literal('un'))("UnNode")
qualifier = Group(linear | unrestricted)("QualifierNode")

pre = Forward()
tybool = Group('bool')("TyBoolNode")
tyend = Group('end')("TyEndNode")
tyqual = Group(qualifier+pre)("TyQualNode")
ty =  Group(tybool | tyend | tyqual)("TyNode")

recv = Group('?'+ty+'.'+ty)("TyRecvNode")
send = Group('!'+ty+'.'+ty)("TySendNode")
pre << Group(recv | send)("PreTypeNode")

var = Group(Word( alphas+"_", alphanums+"_" ))("VariableNode")
vbool =  Group(Literal('true')|Literal('false'))("VBoolNode")
val = Group(var | vbool)("ValueNode")

channel = Group(var)("ChannelNode")

process = Forward()
outputt = Group(channel+val+Literal(".")+process)("OutputNode")
inputt = Group(channel+Literal("(")+val+Literal(")")+Literal(".")+process)("InputNode")
par = Group(Literal("[")+process+Literal("|")+process+Literal("]"))("ParNode")
conditional = Group(Literal("if")+val+Literal("then")+process+Literal("else")+process)("IfNode")
inaction = Group(Literal("0"))("InactionNode")
screst = Group(Literal("(W")+channel+channel+Literal(")")+process)("ScRestNode")
process << Group(outputt | inputt | par | conditional | inaction| screst)("ProcessNode")

class TypeNode():
    def __init__(self, root=None):
        self.root = root
        self.child = []
    
    def addChild(self, c):
        self.child.append(c)
        
def createNode(xmlroot):
    tag = xmlroot.tag
    if tag == "ProcessNode":
        return Process()
    if tag == "InactionNode":
        return Inaction()
    if tag == "InputNode":
        return Input()
    if tag == "OutputNode":
        return Output()
    if tag == "ParNode":
        return Par()
    if tag == "ChannelNode":
        return Channel()
    if tag == "VariableNode":
        return Variable(xmlroot[0].text)
    if tag == "ValueNode":
        return Value()
    if tag == "IfNode":
        return If()
    if tag == "ScRestNode":
        return ScopeRestriction()
    else:
        return None
    
def XMLToTypeTree(xmlroot):
    tree = TypeNode()
    node = createNode(xmlroot)
    
    if xmlroot is None:
        return None
    
    if node is None:
        return None
    
    else:
        tree.root = node
        
    for c in xmlroot:
        #print 'Child:' + str(c)  + ' T:' + str(c.tag) + ' Parent:' + str(tree) + ' T:' + str(xmlroot.tag)
        childTree = XMLToTypeTree(c)
        if childTree:
            tree.addChild(childTree)
    return tree

''' Simple Inorder Printer for the AST '''
def printTree(tree):
    if(tree == None or tree.root == None):
        return
    print tree.root,","
    for c in tree.child:
        print"{"
        printTree(c)
        print"}"

### Test Strings ###
inString = "0"
opString = "x_ v."+inString
ipString = "x (x)."+inString
parString = "["+ipString+"|"+opString+"]"
ifString = "if true then "+opString+" else "+ipString
sdrestString = "(W x1 x2)"+opString


ps = process.parseString(inString)
xmldata = ps.asXML()
print xmldata
root = ET.fromstring(xmldata)
tree = XMLToTypeTree(root)
```
